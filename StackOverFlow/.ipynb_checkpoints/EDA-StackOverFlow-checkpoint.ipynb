# EDA on StactOverflow Data
Source: https://www.ics.uci.edu/~duboisc/stackoverflow/
Dataset of answers posted between February 18 2009 to Jun 7 2009

```python
import pandas as pd
import random
import os
import numpy as np
from tqdm import tqdm_notebook
from collections import Counter

from gensim.models import Word2Vec
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 10.0)
```

```python
path = 'answers.csv'
```

```python
df = pd.read_csv(path)
```

```python
df.columns
```

```python
df = df.drop(labels=['Unnamed: 0'], axis=1)
df = df[np.isfinite(df['i'])]
```

```python
df.columns = ['question_id', 'userId_q', 'score_q', 'time_q',
              'tags', 'views_q', 'answers', 'answer_id',
              'userId_a', 'score_a', 'time_a']
```

```python
len(df)
```

```python
df.head()
```

## What are the most commonly used tags?

```python
tags = list(df['tags'].unique())
all_tags = []
for tag in tqdm_notebook(tags):
    all_tags += tag.split(',')
```

```python
count_tags = Counter(all_tags)
tags = pd.DataFrame([list(count_tags.keys()), list(count_tags.values())])
tags = tags.transpose()
tags.columns = ['tag', 'count']
tags = tags.sort_values(by='count', ascending=False)
```

```python
plt.barh(list(tags['tag'])[:25], list(tags['count'])[:25])
plt.ylabel('Tags Used')
plt.xlabel('Tag Raw-Count')
plt.title('25 Most Popular Tags')
plt.show()
```

From the graph above, we can clearly infer that **C#** is the most commonly used tag followed by **unet**, **java**, **aspnet**, **C++**, **PHP**, **JavaScript**, **Python**, **SQL**, **SQL-Server**

### Let's examine the popularity of some common tags used in Data Science

```python
data_science = ['python', 'r', 'matlab', 'sas', 'excel', 'sql']
x = []
y = []
for tag in data_science:
    index = list(tags['tag']).index(tag)
    x.append(tag)
    y.append(list(tags['count'])[index])
```

```python
plt.bar(x, y)
plt.ylabel('Tag')
plt.xlabel('Tag Raw-Count')
plt.title('Popularity of Data Science Tags')
plt.show()
```

From the graph, **Python** seems to most commonly tag in Data Science, followed by **SQL**, **Excel**, **MATLAB**, **SAS**, **R**.

We should also note here that __python__ is not only used in Data Science, but also in various other fields. This is also one the reasons why number of tags of __python__ is so much more as compared to __R__

## What tags get highest/lowest rating in Questions?

```python
tag_rate = df[['tags', 'score_q']].drop_duplicates().reset_index(drop=True)
tag_rate.head()
```

```python
tagRate = {}
for i in tqdm_notebook(range(len(tag_rate['tags']))):
    for tag in tag_rate['tags'][i].split(','):
        try:
            tagRate[tag].append(tag_rate['score_q'][i])
        except Exception:
            tagRate[tag] = [tag_rate['score_q'][i]]
```

```python
tagRateAvg = {}
for key in list(tagRate.keys()):
    tagRateAvg[key] = np.average(tagRate[key])
```

```python
tagRateAvg = pd.DataFrame([list(tagRateAvg.keys()), list(tagRateAvg.values())])
tagRateAvg = tagRateAvg.transpose()
tagRateAvg.columns = ['tags', 'score_q']
tagRateAvg = tagRateAvg.sort_values('score_q')
```

```python
plt.barh(list(tagRateAvg['tags'])[:25], list(tagRateAvg['score_q'])[:25])
plt.xlabel('Average Scoring')
plt.title('25 Lowest scoring tags (Average)')
plt.ylabel('Tag')
```

From the Graph, we can infer that **personality** based questions achieved the lowest scoring followed by **misv**, **hide-component**, **spyware**, **online-video**, **kthnxbye**, **geekdom**, **human**, **sqlxmlreader**, **dancer**.

We can also note that these tags convey the most general and vague ideas, and if you are familiar with stackover flow platform you must have also noticed that the community is not a fan of vague questions because, they prefer specific questios that they can help with.

```python
plt.barh(list(tagRateAvg['tags'])[-25:], list(tagRateAvg['score_q'])[-25:])
plt.xlabel('Average Scoring')
plt.title('25 Highest scoring tags (Average)')
plt.ylabel('Tag')
```

From the Graph, we can infer that **joeltest** based questions achieved the highest scoring followed by **peopleware**, **queue-theory**, **golf**, **infix**, **arsenal**, **code-challenge**, **black-hat**, **white-hat**, **habits**.

As you can notice, **R** has the least number of questions (20 Questions), from which not even a single question got a negative scoring.

On the other hand, **Python** has a lot more questions (> 2700 Questions) which contains on question with score as high as 40! But also it contains substantial amount of questions with negative overall score.

## Which comminity is the most Active/Inactive in answering a question?

```python
tag_rate = df[['tags', 'time_q', 'time_a']].drop_duplicates().reset_index(drop=True)
tag_rate['diff'] = abs(tag_rate['time_q'] - tag_rate['time_a'])
tag_rate.head()
```

```python
tagRate = {}
for i in tqdm_notebook(range(len(tag_rate['tags']))):
    for tag in tag_rate['tags'][i].split(','):
        try:
            tagRate[tag].append(tag_rate['diff'][i])
        except Exception:
            tagRate[tag] = [tag_rate['diff'][i]]
```

```python
tagRateAvg = {}
for key in list(tagRate.keys()):
    tagRateAvg[key] = np.average(tagRate[key])
```

```python
tagRateAvg = pd.DataFrame([list(tagRateAvg.keys()), list(tagRateAvg.values())])
tagRateAvg = tagRateAvg.transpose()
tagRateAvg.columns = ['tags', 'diff']
tagRateAvg = tagRateAvg.sort_values('diff')
```

```python
plt.barh(list(tagRateAvg['tags'])[:25], list(tagRateAvg['diff'])[:25])
plt.xlabel('Average Time taken to Respond/Answer')
plt.ylabel('Tag')
plt.title('Top-25 Most Active Communities')
plt.show()
```

Surprisingly, **Tetris** community is the most active followed by, **nsbrowser**, **quicksilver**, **no-javascript**, **named-instance**, **inittab**, **gregorian-calender**, **cf8**, **cffeed** and **serving**.

```python
plt.barh(list(tagRateAvg['tags'])[-25:], list(tagRateAvg['diff'])[-25:])
plt.xlabel('Average Time taken to Respond/Answer')
plt.ylabel('Tag')
plt.title('Top-25 Most Inactive Communities')
```

Getting into, inactive communities, **systemaddin** comes first, the reason can be the mis-spelled tag, followed by **livecycle**, **wmi-query**, **jcarousel**, **continuum**, **web-extensibility**, **newtonsoft**, **address-bar**, **metaweblog** and **yuma**.

# Finding Similar tags

```python
tags = []
for tag in list(df['tags']):
    tags.append(tag.split(','))
```

```python
embedding = Word2Vec(tags, sg=1, size=256, min_count=1)
```

```python
X = embedding[embedding.wv.vocab]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
```

```python
fig, ax = plt.subplots()
ax.plot(result[:, 0], result[:, 1], 'o')
ax.set_title('Embeddings - Spread')
plt.show()
```

```python
i = 50

fig, ax = plt.subplots()
ax.set_title('Embeddings - Annotated')
ax.scatter(result[:i, 0], result[:i, 1])

for j, txt in enumerate(list(embedding.wv.vocab)[:i]):
    ax.annotate(txt, result[j])
```

From the above graph we can notice that terms such as javascript,css, django jquery http webrequest etc occur together which actually makes sense as these tags indicate web-development. Similarly terms such as image-processing, core-graphics, design-processing are together indicating image-related topics etc.

By looking deeper into the embedding, we can explore more patterns and relations.

From the graph above, we can notice that r and matlab are closer with machine-learning and data-analysis more as compared to python. This is due to the fact that, Python is used not only for data science, but in various other tasks. If you give a closer look previous graph, one can notice that **Python** lies somewhere in the middle of the plot indicating that Python has a large spread of use and not only limited to data-science or web-development of image-processing.

## Finding the Topic spread of a user's Questions

```python
def plot_tags(tags_df, userId, text):
    for i in range(len(tags_df)):
        tags = tags_df[i].split(',')
        for tag in tags:
            index = word_list.index(tag)
            plt.scatter(result[index, 0], result[index, 1], color='orange')
            plt.annotate(tag, result[index])
    plt.title('{} User ID: {}'.format(text, userId))
    plt.show()
```

```python
grouped_user_id = df.groupby('userId_q')
user_ids_q = df[['userId_q']].drop_duplicates().reset_index(drop=True)
user_ids_q.head(5)
```

```python
index = random.randint(0, len(user_ids_q))
group = grouped_user_id.get_group(user_ids_q['userId_q'][index])[['userId_q', 'tags']].drop_duplicates().reset_index(drop=True)
group
```

```python
plot_tags(group['tags'], group['userId_q'][0], text='Topic Spread')
```

From the graph, the given userID has asked questions on the topics plotted in the graph.
